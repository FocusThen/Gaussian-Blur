```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
image = cv2.imread('../images/brain_MR.jpg')
image_copy = np.copy(image)
image_copy = cv2.cvtColor(image_copy, cv2.COLOR_BGR2RGB)
```

```python
gray = cv2.cvtColor(image_copy, cv2.COLOR_RGB2GRAY)
blur_img = cv2.GaussianBlur(gray, (5,5), 0)

f, (ax1, ax2) = plt.subplots(1, 2, figsize=(20,10))

ax1.imshow(gray, cmap="gray")
ax1.set_title("Original")

ax2.imshow(blur_img, cmap="gray")
ax2.set_title("Blured Image")
```

```python
sobel_x = np.array([[-2, 0,2],
                    [-1, 0, 1],
                    [-2, 0, 2]])
```

```python
filtered_gray = cv2.filter2D(gray, -1, sobel_x)
filtered_blur = cv2.filter2D(blur_img, -1, sobel_x)
```

```python
f, (ax1, ax2) = plt.subplots(1, 2, figsize=(20,10))

ax1.imshow(filtered_gray, cmap="gray")
ax1.set_title("Filtered Gray")

ax2.imshow(filtered_blur, cmap="gray")
ax2.set_title("Filtered Blur")
```
